```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

# Crack 1


```python
def crackdetection(img):
    ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY) #ori value 127
    improved_sp_img = cv2.medianBlur(thresh1, 9)
    minElement = np.amin(improved_sp_img)
    
    if minElement == 255:  #white pixels
        print("Good wall condition")
        
        
    elif minElement == 0:  #black pixels 
        print("Alert: Cracks are detected")
    

    plt.imshow(improved_sp_img), plt.axis('off')
    plt.savefig('crack_1.png')
```

```python
img = cv2.imread('Q3_crack_1.jpg')

crackdetection(img)
```

# Crack 2


```python
def crackdetection(img):
    ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY) #ori value 127
    improved_sp_img = cv2.medianBlur(thresh1, 11)
    kernel = np.ones((5,5),np.uint8)
    improved_sp_img = cv2.morphologyEx(thresh1, cv2.MORPH_OPEN, kernel)
    improved_sp_img = cv2.medianBlur(improved_sp_img, 3)
    minElement = np.amin(improved_sp_img)
    
    if minElement == 255:   #white pixels
        print("Good wall condition")
        
        
    elif minElement == 0:   #black pixels 
        print("Alert: Cracks are detected")

    
    plt.imshow(improved_sp_img), plt.axis('off')
    plt.savefig('crack_2.png')
```

```python
img = cv2.imread('Q3_crack_2.jpg')
crackdetection(img)
```

# Crack 3


```python
def crackdetection(img):
    improved_sp_img = cv2.medianBlur(img,1)
    ret,thresh1 = cv2.threshold(improved_sp_img,70,255,cv2.THRESH_BINARY) #ori value 127
    minElement = np.amin(thresh1)

    
    if minElement == 255:  #white pixels
        print("Good wall condition")
        
        
    elif minElement == 0:  #black pixels 
        print("Alert: Cracks are detected")
    
    plt.imshow(thresh1), plt.axis('off')
```
